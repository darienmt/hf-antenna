```python
import skrf as rf
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import *
rf.stylely()
```

```python
from constants import bands
print(f'Bands count: {len(bands)}')
```

```python
before_tuning = rf.Network('data/nanovna-20250114-023725.s1p', f_unit='mhz')
tuning1=rf.Network('data/nanovna-20250114-151820.s1p', f_unit='mhz')
tuning2 = rf.Network('data/nanovna-20250114-164414.s1p', f_unit='mhz')
tuning3 = rf.Network('data/nanovna-20250114-165341.s1p', f_unit='mhz')
```

```python
def plot_bands(network, title=""):
    fig = plt.figure(figsize=(12,20))
    fig.suptitle(title, fontsize=16)
    for idx, band in enumerate(bands):
        ax = fig.add_subplot(8,2, idx + 1)
        name = band["name"]
        ax.set_title(f'{name} M')
        nw = network[band["freq"]]
        nw.plot_s_vswr(ax = ax)        
        ax.get_legend().remove()
        print(f'{name} M : {nw}')
    
    fig.tight_layout()

plot_bands(before_tuning, "Before tuning")
```
